## PARTE 1 — Vector con distribución de puntos: `np.linspace()`

### 1.1 Sintaxis y ejemplo básico

```python
# np.linspace(inicio, fin, num_puntos)
# → fin SÍ se incluye  |  num_puntos define la cantidad exacta

vector = np.linspace(0, 1, 5)
print('5 puntos entre 0 y 1:')
print(vector)
```

### 1.2 `linspace()` vs `arange()` — comparación directa

```python
print('=== linspace: 5 puntos entre 0 y 1 (incluye el 1) ===')
print(np.linspace(0, 1, 5))

print()
print('=== arange: paso 0.25 entre 0 y 1 (excluye el 1) ===')
print(np.arange(0, 1, 0.25))

print()
print('=== Problema de precisión con arange() (pasos decimales) ===')
print(np.arange(0, 1, 0.1))     # puede generar imprecisiones de punto flotante

print()
print('=== linspace evita ese problema ===')
print(np.linspace(0, 1, 11))    # 11 puntos exactos, sin imprecisión
```

### 1.3 Aplicaciones típicas

```python
# Eje de tiempo: 100 puntos entre 0 y 10 segundos
tiempo = np.linspace(0, 10, 100)
print('Eje temporal (primeros 10 valores):', tiempo[:10])
print(f'Total de puntos: {len(tiempo)}')

print()
# Escala de temperatura para normalización: 0 a 100°C en 11 divisiones
escala_temp = np.linspace(0, 100, 11)
print('Escala de temperatura (11 divisiones):', escala_temp)

print()
# Visualizar función seno sobre el eje temporal
señal = np.sin(np.linspace(0, 2 * np.pi, 100))
plt.figure(figsize=(8, 3))
plt.plot(tiempo, señal, color='#2E75B6', linewidth=2)
plt.title('Señal simulada con linspace (seno)', fontweight='bold')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## PARTE 2 — Matriz identidad: `np.identity()`

### 2.1 Creación y propiedades

```python
# Matriz identidad 3×3: unos en la diagonal, ceros en el resto
identidad = np.identity(3)
print('Matriz identidad 3×3:')
print(identidad)
print()
print(f'Shape: {identidad.shape}')
print(f'dtype: {identidad.dtype}')
```

### 2.2 Propiedad fundamental: multiplicación por identidad

```python
# La identidad es el "1" de la multiplicación matricial
A = np.array([[2, 3], [4, 5]])
I = np.identity(2)

resultado = np.dot(A, I)   # A × I = A

print('Matriz A:')
print(A)
print()
print('Identidad 2×2:')
print(I)
print()
print('A × Identidad = A:')
print(resultado)
print()
print('¿El resultado es igual a A?', np.array_equal(A, resultado))
```

### 2.3 Uso en calibración de sensores

```python
# La identidad como referencia ideal de calibración (sin distorsión)
# 3 sensores × 3 coeficientes de calibración
calibracion_ideal = np.identity(3)
print('Referencia de calibración (sin distorsión):')
print(calibracion_ideal)

# Lecturas crudas de 3 sensores en 3 variables [temp, humedad, presión]
lecturas_crudas = np.array([
    [22.5, 55.0, 1012.0],
    [23.1, 54.2, 1011.5],
    [22.8, 55.5, 1012.3]
])

# Aplicar calibración (identidad no altera)
lecturas_calibradas = np.dot(lecturas_crudas, calibracion_ideal)
print()
print('Lecturas calibradas (idénticas, calibración neutra):')
print(lecturas_calibradas)
```

## PARTE 3 — Matrices aleatorias: `np.random`

### 3.1 `np.random.rand()` — flotantes entre 0 y 1

```python
# Valores flotantes uniformes entre 0.0 y 1.0
aleatoria = np.random.rand(2, 3)
print('Matriz aleatoria flotante (2×3):')
print(aleatoria)
print(f'Min: {aleatoria.min():.4f} | Max: {aleatoria.max():.4f}')
```

### 3.2 `np.random.randint()` — enteros en un rango

```python
# Enteros entre 10 (inclusive) y 100 (excluido)
enteros = np.random.randint(10, 100, size=(2, 3))
print('Matriz aleatoria entera (2×3):')
print(enteros)
```

### 3.3 Semilla de reproducibilidad: `np.random.seed()`

```python
# Sin semilla: resultados cambian en cada ejecución
print('Sin semilla (cambia cada vez):')
print(np.random.randint(0, 10, 5))
print(np.random.randint(0, 10, 5))

print()
# Con semilla: resultados SIEMPRE iguales
print('Con semilla fija (siempre igual):')
np.random.seed(42)
print(np.random.randint(0, 10, 5))
np.random.seed(42)
print(np.random.randint(0, 10, 5))   # idéntico al anterior
```

### 3.4 Escalar rango de `rand()` a medidas reales

```python
# rand() genera entre 0 y 1 → escalar para simular temperaturas entre 15°C y 30°C
np.random.seed(42)
temp_simuladas = np.random.rand(10) * (30 - 15) + 15
print('Temperaturas simuladas (°C):')
print(np.round(temp_simuladas, 2))
print(f'Rango: {temp_simuladas.min():.2f} — {temp_simuladas.max():.2f}°C')
```

## PARTE 4 — Redimensionado de arreglos

### 4.1 `reshape()` — cambiar forma sin alterar datos

```python
# Arreglo de 12 elementos → matriz 3×4
original = np.arange(12)
print('Arreglo original (1D):', original)
print(f'Shape: {original.shape}')

print()
matriz = original.reshape((3, 4))
print('Reshape a (3, 4):')
print(matriz)
print(f'Shape: {matriz.shape}')
```

### 4.2 Combinaciones válidas de reshape

```python
datos = np.arange(12)   # 12 elementos en total

print('(2, 6):'); print(datos.reshape(2, 6))
print()
print('(6, 2):'); print(datos.reshape(6, 2))
print()
print('(1, 12):'); print(datos.reshape(1, 12))
print()
print('(12, 1):'); print(datos.reshape(12, 1))

# ¿Qué pasa si las dimensiones no coinciden?
try:
    datos.reshape(4, 4)   # 4×4 = 16 ≠ 12
except ValueError as e:
    print()
    print(f'❌ Error al reshape(4,4): {e}')
```

### 4.3 `flatten()` y `ravel()` — reducir a 1D

```python
np.random.seed(42)
luminosidad = np.random.randint(200, 801, size=(3, 5))
print('Matriz original (3×5):')
print(luminosidad)

print()
# flatten() → crea una COPIA independiente en 1D
lineal_flatten = luminosidad.flatten()
print('flatten() — copia 1D:', lineal_flatten)

print()
# ravel() → crea una VISTA (más eficiente en memoria)
lineal_ravel = luminosidad.ravel()
print('ravel()   — vista 1D:', lineal_ravel)

# Demostrar diferencia: modificar flatten no altera original
lineal_flatten[0] = 9999
print()
print('¿flatten modifica el original?', luminosidad[0, 0] == 9999)   # False
```

### 4.4 Uso de `-1` en reshape (dimensión automática)

```python
v = np.arange(24)

# -1 le dice a NumPy que calcule esa dimensión automáticamente
print('reshape(4, -1)  → NumPy calcula columnas:')
print(v.reshape(4, -1))   # 24 / 4 = 6 columnas

print()
print('reshape(-1, 6)  → NumPy calcula filas:')
print(v.reshape(-1, 6))   # 24 / 6 = 4 filas
```

## 📋 Resumen de funciones vistas

| Función | Sintaxis | Descripción |
|---------|----------|-------------|
| `np.linspace()` | `np.linspace(ini, fin, n)` | N puntos equiespaciados, **incluye** el fin |
| `np.identity()` | `np.identity(n)` | Matriz identidad n×n (diagonal = 1) |
| `np.random.rand()` | `np.random.rand(f, c)` | Flotantes aleatorios entre 0 y 1 |
| `np.random.randint()` | `np.random.randint(a, b, size)` | Enteros aleatorios entre a y b-1 |
| `np.random.seed()` | `np.random.seed(n)` | Fijar semilla para reproducibilidad |
| `.reshape()` | `arr.reshape(f, c)` | Cambiar forma sin alterar datos |
| `.flatten()` | `arr.flatten()` | Reducir a 1D — genera **copia** |
| `.ravel()` | `arr.ravel()` | Reducir a 1D — genera **vista** |
| `np.column_stack()` | `np.column_stack([a, b, c])` | Combinar vectores como columnas |

> 💡 **`linspace` vs `arange`:** Usa `linspace` cuando necesites un número exacto de puntos. Usa `arange` cuando necesites controlar el paso.

> 💡 **`reshape` regla de oro:** El producto de las nuevas dimensiones **debe** ser igual al total de elementos originales. Usa `-1` para que NumPy calcule automáticamente una dimensión.

> 💡 **Semilla:** Fija siempre `np.random.seed()` en experimentos que necesites reproducir o compartir.
