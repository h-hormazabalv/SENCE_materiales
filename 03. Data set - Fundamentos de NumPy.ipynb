## PARTE 1 — NumPy vs. Listas de Python

### 1.1 ¿Por qué NumPy y no listas?

```python
# Con listas tradicionales: necesitamos un bucle
lista = [1, 2, 3, 4, 5]
doble_lista = [x * 2 for x in lista]
print('Lista duplicada (bucle):', doble_lista)

# Con NumPy: operación directa (vectorización)
arreglo = np.array([1, 2, 3, 4, 5])
doble_numpy = arreglo * 2
print('Arreglo duplicado (NumPy):', doble_numpy)
```

### 1.2 Comparación de rendimiento

```python
tiempo_lista = timeit.timeit('[x*2 for x in range(10000)]', number=1000)
tiempo_numpy = timeit.timeit('import numpy as np; np.arange(10000)*2', number=1000)

print(f'Lista  → {tiempo_lista:.4f} segundos')
print(f'NumPy  → {tiempo_numpy:.4f} segundos')
print(f'NumPy es {tiempo_lista/tiempo_numpy:.1f}x más rápido')
```

## PARTE 2 — Creación de vectores

### 2.1 Vector básico con `np.array()`

```python
# Vector de temperaturas registradas por un sensor
vector = np.array([10, 20, 30, 40])
print('Vector:', vector)

# Explorar propiedades
print()
print(f'ndim  → {vector.ndim}    (número de dimensiones)')
print(f'shape → {vector.shape}  (cantidad de elementos por dimensión)')
print(f'dtype → {vector.dtype}  (tipo de dato)')
print(f'size  → {vector.size}   (total de elementos)')
```

### 2.2 Operaciones vectorizadas

```python
print('Vector original:     ', vector)
print('Sumar 5 a cada elem: ', vector + 5)
print('Multiplicar por 2:   ', vector * 2)
print('Elevar al cuadrado:  ', vector ** 2)
print('Dividir entre 10:    ', vector / 10)
```

### 2.3 Indexación y slicing

```python
print('Primer elemento:      ', vector[0])    # índice 0
print('Último elemento:      ', vector[-1])   # índice -1
print('Elementos 1 al 2:     ', vector[1:3])  # slicing
print('Desde el índice 2:    ', vector[2:])   # desde índice 2
print('Hasta el índice 2:    ', vector[:2])   # hasta índice 2 (excluido)
```

### 2.4 Conversión desde listas

```python
# Lista de mediciones de temperatura recibida como input
lista_temp = [22.5, 23.0, 21.8, 22.3, 24.1, 21.5, 23.7]

# Convertir a arreglo NumPy
temp_array = np.array(lista_temp)
print('Arreglo de temperaturas:', temp_array)
print('dtype:', temp_array.dtype)
```

## PARTE 3 — Creación de matrices

### 3.1 Matriz básica

```python
# Matriz de lecturas: cada fila = 1 medición, columnas = [temperatura, humedad, partículas]
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print('Matriz:')
print(matriz)

print()
print(f'ndim  → {matriz.ndim}     (dimensiones)')
print(f'shape → {matriz.shape}  (filas x columnas)')
print(f'size  → {matriz.size}     (total elementos)')
print(f'dtype → {matriz.dtype}  (tipo de dato)')
```

### 3.2 Operaciones vectorizadas en matrices

```python
print('Multiplicar por 10:')
print(matriz * 10)

print()
print('Sumar filas con distintos valores (broadcasting):')
print(matriz + np.array([[10, 10, 10], [0, 0, 0]]))
```

### 3.3 Acceso a filas y columnas

```python
print('Elemento [0,1] (fila 0, columna 1):', matriz[0, 1])   # → 2
print('Fila completa [0]:                 ', matriz[0])       # → [1 2 3]
print('Columna completa [:, 1]:           ', matriz[:, 1])    # → [2 5]
print('Submatriz filas 0:0, cols 1:3:     ')
print(matriz[0:1, 1:3])
```

## PARTE 4 — Generación automática con `np.arange()`

### 4.1 Sintaxis y ejemplos básicos

```python
# np.arange(inicio, fin, paso)  →  fin es EXCLUIDO

# Ejemplo básico: de 0 a 20 de 5 en 5
rango = np.arange(0, 20, 5)
print('De 0 a 20 paso 5:', rango)

# Secuencia de minutos del 0 al 9
tiempo = np.arange(0, 10, 1)
print('Minutos [0-9]:   ', tiempo)

# Pasos decimales
fracciones = np.arange(0, 1, 0.2)
print('Fracciones:      ', fracciones)

# Verificar tipos de dato
print()
print('dtype enteros:  ', np.arange(0, 5, 1).dtype)
print('dtype decimales:', np.arange(0, 1, 0.1).dtype)
```

### 4.2 Aplicación práctica: eje temporal

```python
# Eje temporal: lecturas cada 30 segundos durante 5 minutos (300 segundos)
eje_tiempo = np.arange(0, 300, 30)
print('Eje temporal (segundos):', eje_tiempo)
print('Total de lecturas:', len(eje_tiempo))
```

## PARTE 5 — Plantillas con `np.zeros()` y `np.ones()`

### 5.1 Matrices de ceros

```python
# Plantilla vacía: 3 filas × 4 columnas
plantilla = np.zeros((3, 4))
print('Plantilla de ceros (3×4):')
print(plantilla)
print('dtype:', plantilla.dtype)

print()
# Plantilla entera
plantilla_int = np.zeros((3, 4), dtype=int)
print('Plantilla entera:')
print(plantilla_int)
```

### 5.2 Vectores y matrices de unos

```python
# Vector de unos: calibración de 5 sensores
calibracion = np.ones(5)
print('Vector de calibración inicial:', calibracion)

# Matriz de unos 2×3
estructura = np.ones((2, 3))
print()
print('Matriz de unos (2×3):')
print(estructura)

# Multiplicar por un escalar
print()
print('Estructura inicializada en 100:')
print(np.ones((2, 3)) * 100)

# Cambiar dtype
print()
print('Matriz entera (3×3):')
print(np.ones((3, 3), dtype=int))
```

### 5.3 Comparación: `zeros`, `ones`, `full`

```python
forma = (2, 3)

print('np.zeros((2,3)):')
print(np.zeros(forma))

print()
print('np.ones((2,3)):')
print(np.ones(forma))

print()
print('np.full((2,3), fill_value=7):')
print(np.full(forma, fill_value=7))
```

## 📋 Resumen de funciones vistas

| Función | Sintaxis | Descripción |
|---------|----------|-------------|
| `np.array()` | `np.array([1,2,3])` | Crear vector o matriz desde lista |
| `.ndim` | `arr.ndim` | Número de dimensiones |
| `.shape` | `arr.shape` | Forma (filas × columnas) |
| `.dtype` | `arr.dtype` | Tipo de dato |
| `.size` | `arr.size` | Total de elementos |
| `np.arange()` | `np.arange(ini, fin, paso)` | Secuencia numérica automática |
| `np.zeros()` | `np.zeros((f, c))` | Matriz llena de ceros |
| `np.ones()` | `np.ones((f, c))` | Matriz llena de unos |
| `np.full()` | `np.full((f,c), val)` | Matriz llena con valor fijo |

> 💡 **Recuerda:** En `np.arange(inicio, fin, paso)`, el valor `fin` **no se incluye** en el resultado.

> 💡 **Recuerda:** Los arreglos NumPy son **homogéneos**: todos los elementos deben ser del mismo tipo de dato.
